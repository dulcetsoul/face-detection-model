<h1> Matplot lib</h1>

```python
import matplotlib.pyplot as plt
from skimage.transform import resize
```

```python
## reading images
img1 = plt.imread("./pikachu.jpg")
img2 = plt.imread("./evie.jpg")
img3 = plt.imread("./charmandar.jpg")
img4 = plt.imread("./togepi.jpg")
```

```python
## showing images
print(plt.imshow(img1))
```

```python
## slicing the image
pikachu = resize(img1, (200, 200))
charmandar = resize(img2,(200,200))
evie = resize(img3,(200,200))
togepi= resize(img4,(200,200))
```

```python
import numpy as np
h1 = np.hstack((pikachu,charmandar))
h2 = np.hstack((togepi,evie))
img = np.vstack((h1,h2))
plt.imshow(img)
```
