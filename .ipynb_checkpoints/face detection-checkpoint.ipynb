```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

img = plt.imread("./friends.jpg")
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
faces = detector.detectMultiScale(img)
faces
```

```python
faces.shape
```

```python
x,y,w,h = faces[0]
```

```python
image = cv2.rectangle(img , (x,y),(x+w,y+h),(255,0,0),3)
```

```python
plt.imshow(image)
```

```python
for face in faces:
    x,y,w,h = face
    image = cv2.rectangle(img , (x,y),(x+w,y+h),(255,0,0),3)
    plt.imshow(image)
```
